```python
# Імпортуеємо бібліотеку для імпорту даних
from pycaret.datasets import get_data
# Імпортуємо наш датасет
dataset = get_data('income')
# Виводимо його форму, сам датасет та наявні поля
dataset.shape
dataset
dataset.keys()
```

```python
# Переводимо поле з булевим типом зі значень 0 та 1 у True та False
dataset["income >50K"] = [i == 1 for i in dataset["income >50K"]]
# Додаємо поле capital-delta замість полів capital-gain та capital-loss
dataset["capital-delta"] = dataset["capital-gain"] - dataset["capital-loss"]

dataset.drop(labels=["capital-gain","capital-loss"], inplace=True, axis="columns")

# Ще раз виводимо датасет
dataset
```

```python
# Імпортуємо бібліотеку для кластерингу 
from pycaret.clustering import *
# Ініціюємо нашу модель для датасету та зазначаємо параметр нормалізації у True та session_id у 123 задля відтворюваності викликів наступних функцій
cl = setup(dataset, normalize = True, session_id = 123)
cl
```

```python
# Обираємо метод найближчих k сусідів та кількість кластерів=4 для нашої моделі
kmeans = create_model('kmeans', num_clusters = 4)
```

```python
# Отримуємо наш результат
kmean_results = assign_model(kmeans)
```

```python
# Виводимо результат кластерізації
kmean_results
```

```python
# Виводимо графік для даної моделі
plot_model(kmeans)
```

```python
# Виводимо графік розподілу для нашої моделі
plot_model(kmeans, plot = 'distribution')
```

```python

# Виводимо графік розподілу за параметром доходу більше 50 тисяч
plot_model(kmeans, plot = 'distribution', feature="income >50K")
```

```python

# Виводимо графік розподілу за параметром віку
plot_model(kmeans, plot = 'distribution', feature="age")
```

```python
# Виводимо графік розподілу за параметром статі
plot_model(kmeans, plot = 'distribution', feature="sex")
```

```python

# Виводимо графік розподілу за параметром відпрацьованих годин на тиждень
plot_model(kmeans, plot = 'distribution', feature="hours-per-week")
```

```python

# Виводимо графік розподілу за параметром раси
plot_model(kmeans, plot = 'distribution', feature="race")
```

```python

# Виводимо графік розподілу за параметром зайнятості
plot_model(kmeans, plot = 'distribution', feature="workclass")
```

```python
# Виводимо графік розподілу за параметром різниці доходу та розходу капіталу
plot_model(kmeans, plot = 'distribution', feature="capital-delta")
```

```python

# Нижче для кожного кластеру виводимо: середній вік, медіанний вік, середню дельту капіталу, медіанну дельту капіталу
for i in range(4):
    cluster = kmean_results[kmean_results["Cluster"] == f"Cluster {i}"]
    print(f"Cluster {i}")
    print(f"  Average age {cluster['age'].mean()}")
    print(f"  Median age {cluster['age'].median()}")
    print(f"  Average capital delta {cluster['capital-delta'].mean()}")
    print(f"  Median capital delta {cluster['capital-delta'].median()}")
```

```python
# Зберігаємо модель
save_model(kmeans, model_name='kmeans_model')
```

Підсумовуючи усі дані, що були виведені раніше (не ураховуючи дані про зайнятість та расу, окскільки вони мало що нам дають з точки зору різниці між кластерому та тому рівні глибини аналізу, котрий ми проводимо у рамках даної лабораторної роботи), можемо підсумувати наступне:
    Кластер 0 складається з дуже малої кількості людей з дуже малим відсотком (5%) людей, котрі заробляють більше 50 тисяч, переважно це чоловіки. Середній вік 41 рік, медіанний - 39 років. Люди з цієї групи мають найменший середній приріст капіталу (313 одиниць)

    Кластер 1 є найчисельнішим, переважно складається з людей, що заробляють менше 50 тисяч, має приблизно порівну жінок і чоловіків при тому, що жінок набагато менше ніж чоловіків, тобто переважна більшість жінок належать саме до цього кластеру (до речі, те саме відноситься до представників чорної раси). Дана група є найбільш молодою: середній вік - 34.5 років, медіанний вік - 31 рік. Має приріст капіталу трохи більший за мінімальний - 350.

    Кластер 2 є відносно малочисельним, переважно складається з людей, що заробляють менше 50 тисяч, приблизно на дві третини складається з чоловіків, є відносно молодою з середнім віком 37 років та медіанним віком 35 років. Люди з цього кластеру майже виключно є найманними працівниками та мають високий відносно попередніх двох кластерів приріст капіталу - 1810 одиниць.

    Кластер 3 у першу чергу характеризується тим, що майже половина людей, що відносяться до нього, має дохід більший за 50 тисяч. Переважна кількість людей цього кластеру є білими чоловіками відносно срашного віку з середнім віком 43.5 років та медіанним віком 42 роки. Даний кластер має набільшу частину людей, що не є найманими працівниками приватного сектору.
