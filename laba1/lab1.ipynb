```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *  
import pickle


bank = get_data("bank")


feature_description = {
    "age": "Вік клієнта",
    "job": "Тип роботи",
    "marital": "Сімейний стан",
    "education": "Рівень освіти",
    "default": "Наявність кредиту в минулому",
    "balance": "Середній баланс на рахунку",
    "housing": "Чи є житловий кредит",
    "loan": "Чи є персональний кредит",
    "contact": "Спосіб контакту",
    "day": "День місяця останнього контакту",
    "month": "Місяць останнього контакту",
    "duration": "Тривалість останнього контакту в секундах",
    "campaign": "Кількість контактів під час кампанії",
    "pdays": "Кількість днів з моменту останнього контакту (999 - не було контакту)",
    "previous": "Кількість контактів до поточної кампанії",
    "poutcome": "Результат попередньої маркетингової кампанії",
    "deposit": "Цільова змінна: підписка на депозит (yes/no)"
}


clf_setup = setup(bank, target='deposit', session_id=123)


best_model = compare_models()


final_model = finalize_model(best_model)


predictions = predict_model(final_model)


model_filename = "bank_model.pkl"
save_model(final_model, model_filename)

print(f"Модель збережено як {model_filename}")
```
