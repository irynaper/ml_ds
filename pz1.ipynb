```python
from pycaret.datasets import get_data
from pycaret.classification import *

#в моєму варіанті 16, мав використовуватись датасет churn, але get_data не може його знайти
data = get_data('us_presidential_election_results')
print(data.head())
#тому як наш датасет містить результати президентських виборів, то це є завданням
#класифікації (win or loose).

#робимо сетап експерименту класифікації з цільовою змінною party_winner
exp = setup(data, target='party_winner', session_id=123)

#Порівнюємо продуктивність різних моделей класифікації і обираємо найкращу модель
best_model = compare_models()

#Робимо прогноз на основі обраної моделі і датасету
predictions = predict_model(best_model, data=data)

#виводимо результат. 
#Результатом коду будуть прогнозовані класи та ймовірності для нашого набору даних
print(predictions.head())
# Save best model
save_model(best_model, 'best_model')
```
