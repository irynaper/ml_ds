```python
from pycaret.classification import *
from pycaret.datasets import get_data

# Load dataset
data = get_data('us_presidential_election_results')

# Initialize setup
clf = setup(data, target='party_winner', ignore_features=['state', 'cycle', 'incumbent_party', 'incumbent_running'], 
            numeric_features=['dem_poll_avg', 'dem_poll_avg_margin'])

# Compare models
best_model = compare_models()

# Save best model
save_model(best_model, 'best_model')
```
