Для лабораторної роботи використовується dataset Titanic. </br>
Опис полів:</br>
    Survived Survived (0 = No; 1 = Yes) </br>
    Pclass Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)</br>
    name Name</br>
    sex Sex</br>
    age Age</br>
    SibSp Number of Siblings/Spouses Aboard</br>
    Parch Number of Parents/Children Aboard</br>
    ticket Ticket Number</br>
    fare Passenger Fare (British pound)</br>
    cabin Cabin</br>
    embarked Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)</br>

```python
from pycaret.datasets import get_data
from pycaret import classification

titanicDataSet = get_data("titanic")
```

Для аналізу dataset Titanic було обрано завдання кластеризації. 

```python
classification.setup(data= titanicDataSet, target='Survived')
model = classification.create_model('lr')
```

```python
classification.evaluate_model(model)
```

```python
classification.save_model(model, 'save_model')
```
