```python
from pycaret.datasets import get_data 
import pandas as pd 
 
pd.set_option('display.max_rows', 200) 
pd.set_option('display.max_column', 2500) 
pd.set_option('display.max_colwidth', 2500) 
 
NBA = get_data('nba') 
 
data = NBA.sample(frac=0.95, random_state=786) 
data_unseen = NBA.drop(data.index) 
NBA.reset_index(inplace=True, drop=True) 
data_unseen.reset_index(inplace=True, drop=True) 
 
 
from pycaret.classification import * 
s = setup(data = NBA,train_size = 0.9,target='TARGET_5Yrs', session_id=123, normalize = True, remove_multicollinearity=True) 
 
 
compare_models() 
dt = create_model('dt') 
dt.feature_importances_ 
xgboost = create_model('xgboost') 

tuned_dt = tune_model(dt) 

dt_bagged = ensemble_model(dt, method = 'Bagging') 
dt_boosted = ensemble_model(dt, method = 'Boosting') 
evaluate_model(xgboost)
```
