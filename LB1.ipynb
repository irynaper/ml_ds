```python
from pycaret.datasets import get_data
df = get_data('forest')
```

```python
X, Y: координати місця пожежі.
month: місяць, коли сталася пожежа.
day: день тижня, коли сталася пожежа.
FFMC: вказує на вологість лісової підстилки.
DMC: показує вологість мохової підстилки.
DC: показує ступінь сушіння підстилки.
ISI: індекс швидкості розповсюдження пожежі.
temp: температура навколишнього середовища.
RH: відносна вологість повітря.
wind: швидкість вітру.
rain: кількість опадів.
area: площа пожежі, яку охоплює вогонь 
```

```python
from pycaret.classification import *  
import pandas as pd  

# Додаємо бінарний клас: пожежа була чи ні  
df['fire'] = (df['area'] > 0).astype(int)  # 1 = пожежа, 0 = немає  

# Налаштування PyCaret  
clf = setup(df, target='fire', session_id=123, ignore_features=['area', 'month', 'day'])  

# Тренування моделі  
best_model = compare_models()  

# Оцінка моделі  
predict_model(best_model)  
```

```python
gbc_model = create_model('gbc')
tuned_gbc = tune_model(gbc_model, optimize='AUC', custom_grid={
    'n_estimators': [100],  
    'learning_rate': [0.05, 0.1],  
    'max_depth': [3, 5],       
    'subsample': [0.7, 0.8],    
    'random_state': [42],
})
print(tuned_gbc)
```

```python
lightgbm_model = create_model('lightgbm')

tuned_lightgbm = tune_model(lightgbm_model, optimize='Accuracy',  custom_grid={
        'feature_fraction': [0.4, 0.5, 0.6],
        'num_leaves': [20, 30, 40],  
        'learning_rate': [0.01, 0.05, 0.1],  
        'lambda_l1': [0.1, 0.5, 1],  
        'lambda_l2': [0.1, 0.5, 1], 
        'min_child_weight': [1, 3, 5] 
    })
```

```python
Найкращі моделі: Моделі "Ada Boost Classifier", "Gradient Boosting Classifier", "Light Gradient Boosting Machine" досягли дуже високих результатів по всіх метриках (accuracy, precision, recall, F1, AUC, Kappa, MCC), що свідчить про їхню високу ефективність.
Найгірші моделі: Моделі, такі як "SVM - Linear Kernel", "Dummy Classifier", "Ridge Classifier" і "Linear Discriminant Analysis" показують значно гірші результати в порівнянні з іншими моделями.
```
