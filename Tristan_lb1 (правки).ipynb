```python
# Получем датасета glass из PyCaret
from pycaret.datasets import get_data
from pycaret.classification import *
data = get_data('glass')
```

Датасет glass содержит данные о химическом составе стекла и их индексе преломления. Задача состоит в том, чтобы на основе химического состава стекла определить его тип. Датасет содержит следующие колонки:

RI: относительный показатель преломления
Na: содержание натрия
Mg: содержание магния
Al: содержание алюминия
Si: содержание кремния
K: содержание калия
Ca: содержание кальция
Ba: содержание бария
Fe: содержание железа
Type: тип стекла (целевая переменная)

```python
#Настраиваем експеримент
clf = setup(data = data, target = 'Type')
#создаём модель (linear regression)
lr = create_model('lr')
#оценитваем модель
evaluate_model(lr)
#Получаем прогноз
predict_model(lr, data=data)
#Сохраняем модель
save_model(lr, 'best_model')
```
