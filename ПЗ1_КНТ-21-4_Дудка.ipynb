```python
from pycaret.datasets import get_data
data = get_data('jewellery')
```

Датасет jewellery включає чотири числові ознаки:
Age — вік особи;
Income — рівень доходу;
SpendingScore — індекс, що відображає схильність до витрат (наприклад, частота чи обсяг покупок);
Savings — обсяг накопичень.
Кожен запис у датасеті відповідає певному клієнту. Наявні характеристики можуть використовуватися для аналізу споживчої поведінки, наприклад, для сегментації клієнтів за допомогою кластеризації. Датасет не містить цільової змінної, тому він є типовим прикладом для навчання моделей без учителя (unsupervised learning), зокрема для алгоритмів, таких як K-середніх або ієрархічна кластеризація.

```python
Ми передбачаємо числову змінну SpendingScore, отже, задача — регресія.
```

```python
from pycaret.regression import *
reg = setup(data=data, target='SpendingScore', session_id=123,
            train_size=0.8, silent=True, verbose=False)
```

### Аналіз метрик моделі

- **MAE** — показує середню абсолютну похибку моделі.
- **MSE** — враховує великі помилки сильніше, ніж MAE.
- **RMSE** — корінь з MSE, інтерпретується в одиницях цільової змінної.
- **R2** — показує, наскільки добре модель пояснює варіацію цільової змінної.
