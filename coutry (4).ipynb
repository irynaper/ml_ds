```python
from pycaret.datasets import get_data
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


import plotly.express as px
!pip install -U kaleido
import kaleido
```

```python
df = get_data("country-data")
```

```python
df.columns
```

***Трішки пошукавши додаткової інформації я знайшов опис для для кожного атрибута поточного датасету***:
* country - Name of the country
* child_mort - Death of children under 5 years of age per 1000 live births
* exports - Exports of goods and services per capita. Given as %age of the GDP per capita
* health - Total health spending per capita. Given as %age of GDP per capita
* imports - Imports of goods and services per capita. Given as %age of the GDP per capita
* Income - Net income per person
* Inflation - The measurement of the annual growth rate of the Total GDP
* life_expec - The average number of years a new born child would live if the current mortality patterns are to remain the same
* total_fer - The number of children that would be born to each woman if the current age-fertility rates remain the same.
* gdpp - The GDP per capita. Calculated as the Total GDP divided by the total population.

```python
corr = df.drop('country', axis=1).corr()
cmap = sns.diverging_palette(250, 15, s=75, l=40,
                            n=9, center="light", as_cmap=True)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, cmap=cmap, mask=mask, center=0)
plt.plot()
```

***Можна побачити, що атрибут `income` має досить значну кореляцію за такими атрибутами як `gdpp`, `life_expec`, `exports`, `total_fer` та `child_mort`. Таким чином можна зробити регресійну модель за цими параметрами. Також можна виконати кластеризацію та подивитися на результат***

```python
from pycaret import clustering
```

```python
df.set_index('country', inplace=True)
df.head(5)
```

```python
set_up = clustering.setup(df, normalize=True, pca=True) 
```

```python
kmeans = clustering.create_model('kmeans',num_clusters=3)

res = clustering.assign_model(kmeans)
res
```

```python
res['Cluster'].value_counts()
```

```python
clustering.evaluate_model(kmeans)
```

```python
res[res['Cluster'] == 'Cluster 0'].head(5)
```

```python
res[res['Cluster'] == 'Cluster 1'].head(5)
```

```python
res[res['Cluster'] == 'Cluster 2'].head(5)
```

```python
df = res
df.reset_index(inplace=True)
df
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=res, x='gdpp', y='income', hue='Cluster')
plt.plot()
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=res, x='child_mort', y='total_fer', hue='Cluster')
plt.plot()
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=res, x='child_mort', y='gdpp', hue='Cluster')
plt.plot()
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=res, x='inflation', y='exports', hue='Cluster')
plt.plot()
```

```python
fig = px.choropleth(df[['country','Cluster']],
                    locationmode = 'country names',
                    locations = 'country',
                    color = df['Cluster'],  
                    color_discrete_map = {'Cluster 0':'Red',
                                        'Cluster 1':'Yellow',
                                        'Cluster 2':'Green'}
                   )
fig.update_geos(fitbounds = "locations", visible = True)
fig.update_layout(legend_title_text = 'Labels',legend_title_side = 'top',title_pad_l = 260,title_y = 0.86)
fig.show(engine = 'kaleido')
```

```python
clustering.save_model(kmeans, model_name='kmeans')
```
